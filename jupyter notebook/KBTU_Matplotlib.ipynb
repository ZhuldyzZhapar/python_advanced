```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5,5,22)
y = x ** 2
z = x ** 3
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,c='violet')
plt.xlabel('This is an X label')
plt.ylabel('This is a square of X')
plt.title('KBTU Squares')
```

```python
plt.plot(x,y,'r--')
plt.xlabel('This is an X label')
plt.ylabel('This is a square of X')
plt.title('KBTU Squares')
```

```python
plt.plot(x,y,'r*-')
plt.xlabel('This is an X label')
plt.ylabel('This is a square of X')
plt.title('KBTU Squares')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g*-')
plt.subplot(2,2,3)
plt.plot(y,y,'g*-')
plt.subplot(2,2,4)
plt.plot(x,x,'green')
```

```python
plt.plot(x,y,'red',label='x**2')
plt.plot(x,z,'green',label='x**3')
plt.legend()
```

```python
fig = plt.figure(figsize=(7,7))
plt.plot(x,y,'red',label='x**2')
plt.plot(x,z,'green',label='x**3')
plt.legend()
```

```python
fig = plt.figure(figsize=(7,7))
plt.plot(x,y,'red',label='x**2')
plt.plot(x,z,'green',label='x**3')
plt.ylim(-200,200)
plt.legend()
```
